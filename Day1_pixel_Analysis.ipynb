```python
import cv2
import numpy as np
import urllib.request
from google.colab.patches import cv2_imshow

print("1.멋진 빨간색 스포츠카 사진을 불러옵니다...")
url='https://images.unsplash.com/photo-1614162692292-7ac56d7f7f1e?ixlib=rb-4.0.3&auto=format&fit=crop&w=400&q=80'
urllib.request.urlretrieve(url, 'my_car.jpg')


img=cv2.imread('my_car.jpg')
cv2_imshow(img)


print("\n🤖 2. AI가 바라보는 이 자동차의 '진짜 스펙' (Shape)")

print(f"사진의 크기(shape):{img.shape}")


print("\n🔢 3. 화면 왼쪽 위 5x5 픽셀의 실제 숫자들 까보기!")
print(img[0:5,0:5,0])

print("\n🔴 4. 빨간색(Red) 필터만 씌워서 보기!")
# 컬러 사진을 B, G, R 세 장의 흑백 필름으로 쪼갭니다.
b,g,r=cv2.split(img)

cv2_imshow(b)

cv2_imshow(r)

cv2_imshow(g)
```

픽셀과 행렬, 0(완전어두움)~255(완전 밝음)
그리고 컬러사진은 파랑,초록,빨강 3장이나 겹쳐있는 구조
OPEN CV에서는 BGR순이다,

cv2.imread()은 사진을 '숫자 행렬'로 변환해서 메모리에 올린다,

img.shape를 찍어보면 (세로길이,가로길이,색상채널)이 나온다.

행렬 슬라이싱을 통해 픽셀 까보기

2. AI의 시선: 사진은 숫자 행렬이다!

우리가 보는 자동차 사진은 0(어두움) ~ 255(밝음)의 숫자로 꽉 채워진 바둑판이다.

img.shape 코드를 찍어보면 (세로 픽셀, 가로 픽셀, 3) 형태로 나오는데, 여기서 3은 파랑(B), 초록(G), 빨강(R) 3개의 색상 채널이 겹쳐져 있다는 뜻이다.

3. 🎨 색상 채널(RGB)의 엄청난 비밀 (오늘의 하이라이트!)

b, g, r = cv2.split(img)를 통해 컬러 사진을 3장의 흑백 필름으로 쪼개보았다.

빨간색 차의 진실: 빨간색 자동차에 Red 채널(r) 필터를 씌우면, 차체 부분이 255(하얀색)에 가깝게 미친 듯이 밝게 빛난다!

하얀색 차의 진실(돌발 상황): 빛의 삼원색 원리 때문에, 하얀색은 Red(255) + Green(255) + Blue(255)가 모두 최대치로 섞여야 나온다. 그래서 하얀색 포르쉐는 R, G, B 어떤 필터를 씌워도 차체가 계속 하얗게 살아남는다!
