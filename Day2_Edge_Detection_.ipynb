```python
import cv2
import urllib.request
from google.colab.patches import cv2_imshow


print("📸 1. 도로 위의 멋진 스포츠카 사진을 불러옵니다...")
url='https://images.unsplash.com/photo-1492144534655-ae79c964c9d7?w=400'
urllib.request.urlretrieve(url, 'day2_car.jpg')
img=cv2.imread('day2_car.jpg')


print("\n⚫ 2. AI의 집중력을 위해 색상을 버리고 흑백(Grayscale)으로 바꿉니다.")

gray_car=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_car)

print("\n🌫️ 3. 노이즈 제거! 가우시안 블러(Gaussian Blur) 필터 씌우기")
#(5,5)는 블러를 먹이는 '돋보기의 크기'
blurred_car=cv2.GaussianBlur(gray_car,(5,5),0)
cv2_imshow(blurred_car)

print("\n✏️ 4. 오늘의 하이라이트! 캐니 에지(Canny Edge)로 자동차 뼈대 스케치하기")

edges=cv2.Canny(blurred_car,50,150)

cv2_imshow(edges)

```

1.가우시안 블러:사진을 살짝 '뿌옇게'만든다.
사람눈에는 흐릿해 보이지만,AI입장에서는 쓸데없는 먼지나 자글자글
한 노이즈가 싹 지워져서 오히려 집중력이 올라간다.

2.캐니에지:색깔은 다버리고 , 오직 배경과 자동차가 구분되는
경계선(윤곽석)만 하얀펜으로 스케치하듯 따낸다.


*전처리의 정석 3콤보! (흑백 -> 블러 -> 윤곽선)
