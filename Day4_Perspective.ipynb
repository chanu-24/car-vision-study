```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

print("🚗 1. 가상의 '삐딱한 번호판' 이미지를 생성합니다. (URL 에러 원천 차단!)")

img=np.zeros((300,600,3),dtype=np.uint8)

skewed_pts=np.array([[100,100],[450,50],[500,200],[80,250]],np.int32)

cv2.fillPoly(img,[skewed_pts],(255,255,255))
cv2.putText(img,"12 GA 3456",(150,150),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0))
cv2_imshow(img)

print("\n✨ 2. 마법의 스캐너 작동 중... (행렬 계산)")

pts1=np.float32([[100,100],[450,50],[500,200],[80,250]])


width,height=400,100

pts2=np.float32([[0,0],[width,0],[width,height],[0,height]])


matrix=cv2.getPespectiveTransform(pts1,pts2)


print("3.반듯하게 펴진 최종 번호판 결과물!")
flat_plate = cv2.warpPerspective(img, matrix, (width, height))
cv2_imshow(flat_plate)



```

블랙박스나 주차장 cctv에 찍힌 자동차 번호판은 항상 비스금하거나 삐뚫어져 있다.
