```python
import torch
import torchvision.models as models


print("📱 1. 가성비 끝판왕 'MobileNetV2' 뇌 소환 중...")
# 사전 학습된 MobileNetV2 로드
mobilenet=models.mobilenet_v2(weights='IMAGENET1K_V1')


print("\n🔍 2. MobileNet의 내부 구조 살짝 엿보기")
# MobileNet은 마지막 층 이름이 'fc'가 아니라 'classifier'입니다.
print(f"기존 판단 층 구조:\n{mobilenet.classifier}")

print("\n✂️ 3. 초경량 '자동차/자전거' 구분 전문가로 수술")
# MobileNetV2의 마지막 층 입력 노드 개수는 1280개입니다.
num_ftrs = mobilenet.classifier[1].in_features


# 마지막 층을 2개 클래스(자동차, 자전거)로 분류하도록 교체
mobilenet.classifier[1] = torch.nn.Linear(num_ftrs, 2)

print(f"\n✅ 수술 완료! 이제 이 뇌는 {num_ftrs}개의 특징을 조합해 2가지를 구분합니다.")
print(f"수정된 판단 층: {mobilenet.classifier[1]}")

# 📊 모델 크기 확인 (파라미터 수)
total_params = sum(p.numel() for p in mobilenet.parameters())
print(f"\n📏 MobileNetV2 총 파라미터 수: {total_params:,}개")
print("(참고: 어제 본 ResNet18보다 훨씬 가볍고 빠릅니다!)")
```

(실전전이학습)
MobileNet은 스마트폰이나 자율주행 임베디드 칩에서도 쌩썡돌아갈 만큽 가볍다.

표준합성곱,분리 합성곱

사용자가 입력하는 커널 크기와 채널수에 따라 연산향과 파라미터수가 얼마나 획기적으로 줄어드는지 확인


구조의 차이:모델마다 마지막 층 이름이 다릅니다,
ResNet은 .fc
MobileNet은 .classifier

-->이걸 잘 찾아내야 수술(전이 학습)이 가능합니다.

2.연산효율: 위 계산기에서 보셨듯이, 커널크기가 3x3일때 분리 합성솝은 일반방식보다 약 8~9배 적은 연산량으로 비슷한 효과
