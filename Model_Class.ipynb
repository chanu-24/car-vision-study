```python
import torch
import torch.nn as nn
import urllib.request
from torchvision import transforms
from google.colab.patches import cv2_imshow
import cv2

url = 'https://raw.githubusercontent.com/ultralytics/yolov5/master/data/images/bus.jpg'
urllib.request.urlretrieve(url, 'bus.jpg')
img_cv2 = cv2.imread('bus.jpg')
img_tensor = transforms.ToTensor()(img_cv2).unsqueeze(0)

class SimpleCarBrain(nn.Module):
    def __init__(self):
        super(SimpleCarBrain, self).__init__()
        # 부품 창고 세팅
        self.conv = nn.Conv2d(3, 16, kernel_size=3, padding=1) # 돋보기 16개 장착!
        self.pool = nn.MaxPool2d(2, 2)                         # 크기 반토막 압축기
        self.relu = nn.ReLU()                                  # 마이너스 제거 필터

    def forward(self, x):

        x = self.conv(x)
        x = self.relu(x)
        x = self.pool(x)
        return x


        # 🧠 2. 설계도로 AI 뇌 생성
model = SimpleCarBrain()
print("✅ AI 뇌 설계도 조립 완료!")
print(model) # 내 뇌 구조가 어떻게 생겼는지 확인!

# ⚡ 3. 사진 통과!
output = model(img_tensor)

print(f"\n🎯 최종 결과 텐서 크기: {output.shape}")
print("➔ (1박스, 16장의 특징 도화지, 세로/2, 가로/2) 로 변신 성공!")
```

돋보기(conv)로 특징을 뽑고, 엑기스(pool)만 남기는 법을 배웠다.
오늘은 이 파편화된 부품들을 하나로 묶어, 진짜 인공지능의 설계도를 완성

AI는 nn.Module이라는 커다란 상자(클래스)안에 순서대로 담아서 관리

__init__:(부품준비):돋보기,필터,압축기를 창고에서 꺼내온다
foward:(전달):사진이 들어오면 "1번 돋보기 통과!->2번 압축기 통과!
