```python
import cv2
import torch
import urllib.request
from torchvision import transforms
from google.colab.patches import cv2_imshow

print("🚗 1. 바쁘다 바빠! 절대 안 막히는 노란색 페라리 링크로 직행합니다 ㅋㅋㅋ")
# 403 에러 방지용 헤더 추가 (이러면 웹사이트가 로봇인 줄 모르고 그냥 사진을 줍니다!)
url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/b/ba/Yellow_Ferrari_458_Italia.jpg/400px-Yellow_Ferrari_458_Italia.jpg'
req = urllib.request.Request(url, headers={'User-Agent': 'Mozilla/5.0'})

with urllib.request.urlopen(req) as response:
    with open('my_car.jpg', 'wb') as out_file:
        out_file.write(response.read())

# OpenCV로 읽기 (사람의 방식)
img_cv2 = cv2.imread('my_car.jpg')
cv2_imshow(img_cv2)

print("\n📊 2. OpenCV 사진 스펙 (사람 모드)")
print(f"모양(Shape): {img_cv2.shape} ➔ (세로, 가로, 색상) 순서!")

print("\n✨ 3. AI 뇌로 전송! (텐서 변환)")
transform = transforms.ToTensor()
img_tensor = transform(img_cv2)

print("\n🧠 4. PyTorch 텐서 스펙 (AI 모드)")
print(f"모양(Shape): {img_tensor.shape} ➔ (색상, 세로, 가로) 순서로 뒤집힘!")

```

자동차 사진은 0~255의 숫자로 된 numpy행렬

인공지능 뇌 (pyTorch)는 이 숫자를 그대로 주면 뱉어버린다.

AI는 오직 텐서(TENSOR)라는 특수한 형태의 숫자뭉치만 소화할수 있다.

가장 큰 차이점은 데이터를 읽는 순서이다.

OPEN CV(사람의 방식):HWC(세로,가로,색상)

PyTorch(AI의 방식):CHW(색상 세로 가로)
