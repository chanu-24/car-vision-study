```python
import torch
import torch.nn as nn

# 1. 어제 만든 그 뇌(Model) 소환
model=nn.Sequential(nn.Flatten(),nn.Linear(3*32*32,2))

print("💾 1. 공부한 뇌 저장하기")
# 'model.state_dict()'는 AI 뇌세포의 모든 수치(가중치)를 딕셔너리 형태로 담고 있습니다.
torch.save(model.state_dict(),'my_car_model.pth')
print("✅ 'my_car_model.pth' 파일로 저장 완료!")


print("\n📦 2. 저장된 뇌 다시 불러오기 (새로운 노트북이라고 가정)")
new_nodel=nn.Sequential(nn.Flatten(),nn.Linear(3*32*32,2))
new_model.load_state_dict(torch.load('my_car_model.pth'))

print("\n🧐 3. 실전 테스트(Inference) 모드 돌입!")
# 💡 [필수] 추론 전에는 반드시 eval()을 선언해야 합니다! (배치 정규화, 드롭아웃 등 정지)
new_model.eval()

#가짜 자동차 사진 1장
test_image=torch.randn(1,3,32,32)


# 💡 [필수] 추론할 때는 기울기 계산(오답 노트)을 꺼서 속도를 높이고 메모리를 아낍니다.
with torch.no_grad():
    prediction = new_model(test_image)

# 최종 결과 확인 (어떤 숫자가 더 큰가?)
predicted_class = torch.argmax(prediction, dim=1).item()
classes = ['자전거', '자동차']

print(f"🎯 AI의 최종 판정: [ {classes[predicted_class]} ] 입니다!")
```

추론:학습 모드를 끄고,오답노트를 만들지 않는 상태에서 AI에게 정답만 물어보는 과정

임계값:AI는 정답을 말할때 "이건 100%자동차야"라고 하지 않습니다.
"85%확률로 자동차인것 같아"라고 하죠. 이떄 우리가 "80% 넘으면 자동차로 인정해줄게"라고 정하는 기준선이 바로 임계값입니다.

실제 자율주행AI가 물체를 판단할떄 , 이 임계값이 너무 낮으면 헛것을 보고, 너무 높으면 눈앞의 차를 무시하게 됩니다.
