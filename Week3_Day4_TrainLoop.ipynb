```python
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

print("🏭 1. AI 공장 가동 준비 완료! (데이터, 모델, 채점관 세팅)")
# ① 데이터 트럭 준비 (100장, 16장씩 배달)
X_fake = torch.randn(100, 3, 32, 32)
y_fake = torch.randint(0, 2, (100,))
dataloader = DataLoader(TensorDataset(X_fake, y_fake), batch_size=16, shuffle=True)

# ② 아주 간단한 뇌(Model) 준비 (입력 크기를 3*32*32로 맞춤)
model = nn.Sequential(
    nn.Flatten(),
    nn.Linear(3 * 32 * 32, 2) # 자동차냐 자전거냐!
)

# ③ 채점관(Loss)과 과외선생님(Optimizer) 세팅
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01) # 0.01의 최적 학습률!


print("\n🔥 2. 드디어 학습 시작! (Total Epoch: 5)")
# 딥러닝의 꽃, 그랜드 루프 시작!
epochs = 5

# 에포크: 전체 100장 데이터를 몇 번 반복해서 공부할 것인가?
for epoch in range(epochs):
    total_loss = 0 # 이번 에포크의 평균 오차를 구하기 위한 변수

    # 데이터로더 트럭에서 16장씩 꺼내오기!
    for images, labels in dataloader:

        # --------------------------------------------------------
        # 💡 [핵심] 파이토치 개발자가 평생 쓰는 마의 5단계 공식!
        # --------------------------------------------------------
        optimizer.zero_grad()           # 1. 뇌세포 초기화 (이전의 혼난 기억 지우기)
        outputs = model(images)         # 2. 시험 치기 (예측값 뽑기)
        loss = criterion(outputs, labels) # 3. 채점하기 (오차 계산)
        loss.backward()                 # 4. 오답 노트 작성 (어떤 뇌세포가 틀렸나 역추적)
        optimizer.step()                # 5. 뇌세포 업데이트 (진화!)
        # --------------------------------------------------------

        total_loss += loss.item()

    # 1개 에포크가 끝날 때마다 평균 오차(Loss) 출력
    avg_loss = total_loss / len(dataloader)
    print(f"✅ Epoch [{epoch+1}/{epochs}] ➔ 현재 오차(Loss) 점수: {avg_loss:.4f}")

print("\n🏆 학습 완료! AI가 정답에 점점 가까워졌습니다!")
```
